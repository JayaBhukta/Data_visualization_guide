```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import chardet
```

```python
with open('/content/Car details v3.csv','rb') as f:
  encoding=chardet.detect(f.read())['encoding']
df=pd.read_csv('/content/Car details v3.csv',encoding=encoding)
df.head()
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(x='year',y='selling_price',data=df)
plt.show()
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
sns.lineplot(x='year',y='selling_price',data=df,hue='fuel')
plt.subplot(1,2,2)
sns.lineplot(x='year',y='selling_price',data=df,hue='transmission')
plt.show()
plt.figure(figsize=(18,15))
plt.subplot(2,1,1)
sns.lineplot(x='year',y='km_driven',data=df)
plt.subplot(2,1,2)
sns.lineplot(x='year',y='km_driven',data=df,hue='fuel')
plt.show()
plt.figure(figsize=(18,15))
plt.suptitle('Seats category')
plt.subplot(2,1,1)
sns.lineplot(x='year',y='seats',data=df)
plt.subplot(2,1,2)
sns.lineplot(x='year',y='seats',data=df,hue='fuel')
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.suptitle('owner category')
plt.subplot(1,2,1)
plt.pie(df['owner'].value_counts(),autopct='%1.2f%%')
plt.legend(loc='upper left',labels=df['owner'].unique())
plt.subplot(1,2,2)
sns.countplot(x='owner',data=df,palette='Set2')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title('Transmission category based on fuel category')
sns.countplot(x='transmission',data=df,palette='Set2',hue='fuel')
plt.subplot(1,2,2)
plt.pie(df['transmission'].value_counts(),labels=df['transmission'].unique(),autopct='%1.2f%%',colors=['magenta','orange'])
plt.title('Transmission category')
plt.legend(loc='lower left')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title('Fuel category')
plt.pie(df['fuel'].value_counts(),labels=df['fuel'].unique(),autopct='%1.2f%%')
plt.legend(loc='upper right')
plt.subplot(1,2,2)
plt.title('seller category based on fuel')
sns.countplot(x='seller_type',data=df,palette='magma',hue='fuel')
plt.show()
```

```python
plt.figure(figsize=(13,4))
plt.title('selling price vs milege based on fuel category')
sns.scatterplot(x='mileage',y='selling_price',data=df,hue='fuel')
plt.show()
plt.figure(figsize=(13,4))
plt.title('selling price vs engine based on fuel category')
sns.scatterplot(x='engine',y='selling_price',data=df,palette='Set2',hue='fuel')
plt.show()
plt.figure(figsize=(13,4))
plt.subplot(1,2,1)
sns.scatterplot(x='km_driven',y='selling_price',data=df,palette='coolwarm',hue='fuel')
plt.subplot(1,2,2)
sns.scatterplot(x='seats',y='selling_price',data=df)
plt.show()
```
